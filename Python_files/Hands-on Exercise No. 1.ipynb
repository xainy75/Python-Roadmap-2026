✅ Task 1: Salary Tax Calculation (Conditional Statements)

```python
salary = float(input("Enter Your Salary: "))
if salary < 30000:
    tax_rate = 0.05
elif 30000 <= salary <= 70000:
    tax_rate = 0.15
else:
    tax_rate = 0.25

tax = salary * tax_rate
final_salary = salary - tax

print("Salary: ", salary)
print("Tax_rate: ", tax_rate * 100, "%")
print("Tax Amount: ", tax)
print("Final Salary: ", final_salary)

```

✅ Task 2: List → Dictionary Transformation

```python
words = ["apple", "banana", "kiwi", "cherry", "mango"]
word_length_dict = {}
for word in words:
    word_length_dict[word] = len(word)
print(word_length_dict)

```

✅ Task 3: Chat System (OOP Implementation)

```python
class User:
    def __init__(self, name):
        self.name = name

class Message:
    def __init__(self, sender, content):
        self.sender = sender
        self.content = content

class ChatRoom:
    def __init__(self):
        self.users = []
        self.messages = []

    def join(self, user):
        self.users.append(user)
        print(user.name, "joined the chat room.")

    def leave(self, user):
        self.users.remove(user)
        print(user.name, "left the chat room.")

    def send_message(self, sender, content):
        message = Message(sender, content)
        self.messages.append(message)
        print(sender.name, "sent a message:", content)

    def show_chat(self):
        print("\nChat history:")
        for msg in self.messages:
            print(f"{msg.sender.name}: {msg.content}")

# testing the chat room

user1 = User("Ali")
user2 = User("Sara")

chat = ChatRoom()

chat.join(user1)
chat.join(user2)

chat.send_message(user1, "Hello!")
chat.send_message(user2, "Hi Ali!")

chat.show_chat()

chat.leave(user2)
```
